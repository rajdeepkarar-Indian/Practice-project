```python
import pandas as pd
```

```python
cust=pd.read_csv("Customers_Data.csv",parse_dates=["dob", "join_date"])
acc=pd.read_csv("Accounts_Data.csv",parse_dates=["open_date", "close_date"])
txn=pd.read_csv("Transactions_Data.csv",parse_dates=["txn_date"])
```

# Zone wise avg transaction,number of customer,account type

```python
cust=df.groupby(['region','occupation'])['customer_id'].count().reset_index()
cust
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
sns.barplot(data=cust,x='region',y='customer_id',hue='occupation')
```

```python
trans=df.groupby(['region','account_type'])['txn_id'].count().reset_index()
trans
```

```python
sns.barplot(data=trans,x='region',y='txn_id',hue='account_type')
```

```python
trans=df.groupby(['region','account_type'])['amount'].mean().reset_index()
trans
```

```python
sns.barplot(data=trans,x='region',y='amount',hue='account_type')
```
